```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F

from sklearn.model_selection import train_test_split
```

```python
class Model(nn.Module):
    
    def __init__(self, in_features=4, h1=8, h2=8, out_features=3):
        super().__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, out_features)
       
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x
```

```python
torch.manual_seed(32)

model = Model()
```

```python
df = pd.read_csv('../data/pt_data/iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,7))
fig.tight_layout()

plots = [(0,1),(2,3),(0,2),(1,3)]
colors = ['b', 'r', 'g']
labels = ['Iris setosa','Iris virginica','Iris versicolor']

for i, ax in enumerate(axes.flat):
    for j in range(3):
        x = df.columns[plots[i][0]]
        y = df.columns[plots[i][1]]
        ax.scatter(df[df['target']==j][x], df[df['target']==j][y], color=colors[j])
        ax.set(xlabel=x, ylabel=y)

fig.legend(labels=labels, loc=3, bbox_to_anchor=(1.0,0.85))
plt.show()
```

```python
# Split data into training and test sets
X = df.drop('target', axis=1).values
y = df['target'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)

# Convert to tensors
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

# Set Loss function
criterion = nn.CrossEntropyLoss()    # Cross Entropy because multiclass classification

# Choose an optimzier and learning rate
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
model.parameters
```

```python
# Choose number of epochs
epochs = 200
losses = []

# Train the model and record loss
for i in range(epochs):
    # Forward pass and get a prediction
    y_pred = model.forward(X_train)
    
    # Calculate the loss/error
    loss = criterion(y_pred, y_train)
    losses.append(loss)
    
    # Print out information for every 20 epochs
    if i%20==0:
        print(f"epoch {i} and loss is: {loss}")
    
    # BACKPROPAGATION
    optimizer.zero_grad()     # Find where gradients is zero
    loss.backward()
    optimizer.step()
```

```python
# Plot the loss/error
plt.plot(range(epochs), losses)
plt.ylabel('LOSS - CROSS ENTROPY')
plt.xlabel('Epoch')
```

```python
# Evaluate Performance
with torch.no_grad():    # Deactivates gradient process, turns off backpropagation
    y_eval = model.forward(X_test)
    loss = criterion(y_eval, y_test)    
```

```python
loss # Similiar to our trained model loss - GOOD SIGN
```

```python
correct = 0

with torch.no_grad():
    for i,data in enumerate(X_test):
        y_val = model.forward(data)
        print(f"Datapoint {i+1}. {str(y_val)} Actual Class: {y_test[i]}")
        
        if y_val.argmax().item() == y_test[i]:
            correct += 1

print(f"\nCorrect Classifications: {correct}")
```
